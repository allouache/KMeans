```python
k = 3  
centers = []

for i in range(k):
    centroids = np.random.randint(0, data.shape[0])
    centers.append(data[centroids])
centers = np.array(centers)
print(centers)

def distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2, axis=1))


max_iterations = 100
for i in range(max_iterations):
    distances = np.array([distance(data, centroid) for centroid in centers])
    labels = np.argmin(distances, axis=0)
    print(distances)

    new_centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])

    
    if np.all(centers == new_centroids):
        break
    centers = new_centroids

plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='inferno')
plt.scatter(centers[:, 0], centers[:, 1], c='purple', s=200, alpha=0.5,marker='*')
plt.show()
```
